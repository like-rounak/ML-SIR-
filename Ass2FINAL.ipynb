

### **Step 1: Setup and Import Libraries**

This first cell imports all the necessary packages we'll use for loading data, building models, evaluating them, and plotting results.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine, load_digits
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_curve, auc
from itertools import cycle

%matplotlib inline
```

## **Part A: Wine Dataset Classification**

### **Step 2: Load and Prepare the Wine Dataset**

Here, we load the Wine dataset from scikit-learn and separate it into features (`X`) and the target variable (`y`).

```
wine = load_wine()
X_wine = wine.data
y_wine = wine.target
```

### **Step 3: Model Training and Evaluation (Without PCA)**

We will now train and test our classifiers on the original data. We'll loop through different train-test splits (50:50, 60:40, 70:30, 80:20) and evaluate each model.

```
# Create a dictionary to store all our results
wine_results = []

# Define the train-test splits we want to try
test_sizes = [0.5, 0.4, 0.3, 0.2]

# Define the classifiers
classifiers = {
    'SVM-Linear': SVC(kernel='linear', probability=True, random_state=42),
    'SVM-Polynomial': SVC(kernel='poly', probability=True, random_state=42),
    'SVM-RBF': SVC(kernel='rbf', probability=True, random_state=42),
    'SVM-Sigmoid': SVC(kernel='sigmoid', probability=True, random_state=42),
    'MLP': MLPClassifier(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42)
}

# Loop over each train-test split
for size in test_sizes:
    print(f"======================================================")
    print(f"   RESULTS FOR TRAIN-TEST SPLIT: {int((1-size)*100)}:{int(size*100)}   ")
    print(f"======================================================")

    X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, test_size=size, random_state=42, stratify=y_wine)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Loop over each classifier
    for name, clf in classifiers.items():
        print(f"\n----- Classifier: {name} -----")

        clf.fit(X_train_scaled, y_train)
        y_pred = clf.predict(X_test_scaled)

        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='macro')
        recall = recall_score(y_test, y_pred, average='macro')
        f1 = f1_score(y_test, y_pred, average='macro')

        wine_results.append({
            'Split': f"{int((1-size)*100)}:{int(size*100)}",
            'Classifier': name,
            'PCA': 'No',
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1-Score': f1
        })

        print(f"Accuracy: {accuracy:.4f}")
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred, target_names=wine.target_names))

        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(6, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=wine.target_names, yticklabels=wine.target_names)
        plt.title(f'Confusion Matrix for {name} ({int((1-size)*100)}:{int(size*100)} Split)')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.show()

        if name == 'MLP':
            plt.figure(figsize=(8, 4))
            plt.plot(clf.loss_curve_)
            plt.title(f"Loss Curve for MLP ({int((1-size)*100)}:{int(size*100)} Split)")
            plt.xlabel("Epochs")
            plt.ylabel("Loss")
            plt.show()
```

### **Step 4: Parameter Tuning with GridSearchCV**

Now we'll try to improve our MLP classifier by finding the best parameters using `GridSearchCV`. We will do this for the 80:20 split, which is the most common.

```
print("======================================================")
print("   PARAMETER TUNING (GRIDSEARCHCV) FOR 80:20 SPLIT   ")
print("======================================================")

X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, test_size=0.2, random_state=42, stratify=y_wine)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# MLP Tuning
mlp_params = {
    'hidden_layer_sizes': [(50,), (100,)],
    'learning_rate_init': [0.001, 0.01],
    'momentum': [0.9, 0.95]
}

mlp_grid = GridSearchCV(MLPClassifier(max_iter=1000, random_state=42), param_grid=mlp_params, cv=3, n_jobs=-1, scoring='accuracy')
mlp_grid.fit(X_train_scaled, y_train)
best_mlp = mlp_grid.best_estimator_

print("\n----- Best MLP Classifier after Tuning -----")
print(f"Best Parameters: {mlp_grid.best_params_}")

y_pred_mlp = best_mlp.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred_mlp)
precision = precision_score(y_test, y_pred_mlp, average='macro')
recall = recall_score(y_test, y_pred_mlp, average='macro')
f1 = f1_score(y_test, y_pred_mlp, average='macro')

wine_results.append({
    'Split': '80:20',
    'Classifier': 'MLP-Tuned',
    'PCA': 'No',
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1
})

print(f"Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_mlp, target_names=wine.target_names))
cm = confusion_matrix(y_test, y_pred_mlp)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=wine.target_names, yticklabels=wine.target_names)
plt.title('Confusion Matrix for Tuned MLP')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

plt.figure(figsize=(8, 4))
plt.plot(best_mlp.loss_curve_)
plt.title("Loss Curve for Tuned MLP")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

### **Step 5: Applying Principal Component Analysis (PCA)**

Let's reduce the number of features using PCA and see how our models perform. We'll choose the number of components that explain 95% of the data's variance.

```
scaler = StandardScaler()
X_wine_scaled = scaler.fit_transform(X_wine)

pca = PCA(n_components=0.95)
X_wine_pca = pca.fit_transform(X_wine_scaled)

print(f"Original number of features: {X_wine.shape[1]}")
print(f"Reduced number of features after PCA: {X_wine_pca.shape[1]}")

# We will now re-run the models on the PCA-reduced data for the 80:20 split

print("\n======================================================")
print("   RESULTS FOR 80:20 SPLIT WITH PCA   ")
print("======================================================")

X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_wine_pca, y_wine, test_size=0.2, random_state=42, stratify=y_wine)

for name, clf in classifiers.items():
    print(f"\n----- Classifier: {name} with PCA -----")

    clf.fit(X_train_pca, y_train_pca)
    y_pred_pca = clf.predict(X_test_pca)

    accuracy = accuracy_score(y_test_pca, y_pred_pca)
    precision = precision_score(y_test_pca, y_pred_pca, average='macro')
    recall = recall_score(y_test_pca, y_pred_pca, average='macro')
    f1 = f1_score(y_test_pca, y_pred_pca, average='macro')

    wine_results.append({
        'Split': '80:20',
        'Classifier': name,
        'PCA': 'Yes',
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1
    })

    print(f"Accuracy: {accuracy:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test_pca, y_pred_pca, target_names=wine.target_names))

    cm = confusion_matrix(y_test_pca, y_pred_pca)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=wine.target_names, yticklabels=wine.target_names)
    plt.title(f'Confusion Matrix for {name} with PCA')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
```

### **Step 6: ROC Curve and AUC**

The Receiver Operating Characteristic (ROC) curve is a great way to visualize classifier performance. We'll plot it for our best models on the 80:20 split.

```
X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, test_size=0.2, random_state=42, stratify=y_wine)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

y_test_bin = label_binarize(y_test, classes=np.unique(y_wine))
n_classes = y_test_bin.shape[1]

best_classifiers = {
    'SVM-RBF': SVC(kernel='rbf', probability=True, random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Tuned MLP': best_mlp
}

plt.figure(figsize=(10, 8))

for name, clf in best_classifiers.items():
    clf.fit(X_train_scaled, y_train)
    y_score = clf.predict_proba(X_test_scaled)

    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    plt.plot(fpr["micro"], tpr["micro"],
             label=f'{name} (AUC = {roc_auc["micro"]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-class ROC Curve for Wine Dataset')
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

### **Step 7: Final Performance Comparison for Wine Dataset**

Let's put all the results into a single table to easily compare the performance of all our models.

```
wine_df = pd.DataFrame(wine_results)
wine_df.round(4).sort_values(by='Accuracy', ascending=False)
```

Plotting the ROC curve for the Digits dataset.

```
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits, test_size=0.2, random_state=42, stratify=y_digits)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

y_test_bin = label_binarize(y_test, classes=np.unique(y_digits))
n_classes = y_test_bin.shape[1]

best_classifiers_digits = {
    'SVM-RBF': SVC(kernel='rbf', probability=True, random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'MLP': MLPClassifier(max_iter=1000, random_state=42)
}

plt.figure(figsize=(10, 8))

for name, clf in best_classifiers_digits.items():
    clf.fit(X_train_scaled, y_train)
    y_score = clf.predict_proba(X_test_scaled)

    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    plt.plot(fpr["micro"], tpr["micro"],
             label=f'{name} (AUC = {roc_auc["micro"]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-class ROC Curve for Digits Dataset')
plt.legend(loc="lower right")
plt.grid()
plt.show()
```
